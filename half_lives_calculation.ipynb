```python
#import the required libraries 
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

```python
file_path = 'DecayTimecourse.txt'# Define the file path
df = pd.read_csv(file_path, sep='\t', header=1)# Read the dataset from the file
df
```

```python
time_points = df.columns[1:10]  # Extract time points from the DataFrame
time_points
```

```python
time_course_1 = df.iloc[:, :10]# Select columns for the first time course
time_course_1
```

```python
drop_NaN_1 = time_course_1.dropna(subset=[time_course_1.columns[1]])# Drop rows with NaN in the second column
drop_NaN_1
```

```python
final_1 = drop_NaN_1.interpolate(method = 'linear', limit_direction = 'both')# Interpolate to fill remaining NaNs
final_1
```

```python
time_course_2 = df.iloc[:, [0] + list(range(10, 19))]# Select columns for the second time course
time_course_2
```

```python
drop_NaN_2 = time_course_2.dropna(subset=[time_course_2.columns[1]])# Drop rows with NaN in the second column
drop_NaN_2
```

```python
final_2 = drop_NaN_2.interpolate(method = 'linear', limit_direction = 'both')# Interpolate to fill remaining NaNs
final_2
```

```python
time_course_3 = df.iloc[:, [0] + list(range(19, 28))]# Select columns for the third time course
time_course_3
```

```python
drop_NaN_3 = time_course_3.dropna(subset=[time_course_3.columns[1]])# Drop rows with NaN in the second column
drop_NaN_3
```

```python
final_3 = drop_NaN_3.interpolate(method = 'linear', limit_direction = 'both')# Interpolate to fill remaining NaNs
final_3
```

```python
def half_life(time, intensity): # Define the half-life calculation function
    decay_func = lambda t, a, b: a * np.exp(-b * t)# Exponential decay function
    try:
        params, _ = curve_fit(decay_func, time, intensity, maxfev=10000)# Fit the exponential decay model
        hl = np.log(2) / params[1]# Calculate half-life: ln(2) / decay constant
        return hl
    except:
        return np.nan

def calculate_half_lives(time_course): # Function to calculate half-lives for a time course
    half_lives = []
    for index, row in time_course.iterrows():
        intensity = row[1:].dropna().values # Extract intensity values
        time = np.array(time_points[:len(intensity)], dtype=float) # Match time points with intensities
        hl = half_life(time, intensity) # Calculate half-life
        half_lives.append(hl)
    return half_lives
```

```python
half_lives_1 = calculate_half_lives(final_1) # Half-lives for the first time course
half_lives_1
```

```python
half_lives_df = pd.DataFrame({'YORF': final_1['YORF'], 'half_life_1': half_lives_1}) # DataFrame for the first time course
half_lives_df
```

```python
half_lives_2 = calculate_half_lives(final_2) # Half-lives for the second time course
half_lives_2
```

```python
half_lives2_df = pd.DataFrame({'YORF': final_2['YORF'], 'half_life_2': half_lives_2}) # DataFrame for the second time course
half_lives2_df
```

```python
half_lives_3 = calculate_half_lives(final_3) # Half-lives for the third time course
half_lives_3
```

```python
half_lives3_df = pd.DataFrame({'YORF': final_3['YORF'], 'half_life': half_lives_3}) # DataFrame for the third time course
half_lives3_df
```

```python
merged_df = pd.merge(half_lives_df, half_lives2_df, on='YORF', how='outer', suffixes=('_1', '_2'))# Merge first two DataFrames
merged_df = pd.merge(merged_df, half_lives3_df, on='YORF', how='outer') # Merge with the third DataFrame
merged_df.rename(columns={'half_life': 'half_life_3'}, inplace=True) # Rename column for clarity
merged_df
```

```python
merged_df.head(20)
```

```python
merged_df['average_half_life'] = merged_df[['half_life_1', 'half_life_2', 'half_life_3']].mean(axis=1) # Calculate average half-life
merged_df
```

```python
merged_df.head(10) 
```

```python
sorted_df = merged_df.sort_values(by='average_half_life', ascending=False)# Sort the merged DataFrame by 'average_half_life' in descending order
top_10_percent_count = int(len(sorted_df) * 0.1)# Calculate the number of genes representing the top 10%
top_10_percent_genes = sorted_df.head(top_10_percent_count)# Select the top 10% genes based on the highest average half-lives
top_10_percent_genes
```

```python
top_10_percent_genes.to_csv('top_10.csv', index=False)# Save the top 10% genes in a csv
```

```python
sorted_df = merged_df.sort_values(by='average_half_life', ascending=True)# Sort the merged DataFrame by 'average_half_life' in ascending order
bottom_10_percent_count = int(len(sorted_df) * 0.1)# Calculate the number of genes representing the bottom 10%
bottom_10_percent_genes = sorted_df.head(bottom_10_percent_count)# Select the bottom 10% genes based on the highest average half-lives
bottom_10_percent_genes
```

```python
bottom_10_percent_genes.to_csv('bottom_10.csv', index=False)# Save the bottom 10% genes in a csv
```
